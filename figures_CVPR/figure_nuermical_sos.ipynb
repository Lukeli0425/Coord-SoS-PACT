```python
import sys

sys.path.append('../')

import numpy as np
from matplotlib import pyplot as plt
from tqdm import tqdm

from utils.dataio import *
from utils.visualization import *

plt.rcParams.update({'font.weight': 'semibold'})

DATA_PATH = '../data/'
```

### Load data

```python
config = load_config('../config.yaml')
```

```python
norm_sos = Normalize(vmax=1610, vmin=1490)
```

```python
fig = plt.figure(figsize=(15, 6))

for idx in range(5):
    tps = config[f'numerical {idx}']
    sos = load_mat(os.path.join(DATA_PATH, tps['SOS']))
    
    ax = plt.subplot(1,5,idx+1)
    plt.imshow(sos, cmap='magma', norm=norm_sos)
    plt.title(f'SOS {idx+1}', fontsize=16, fontweight='semibold')
    plt.axis('off')
    if idx == 4:
        cax = fig.add_axes([ax.get_position().x1+8e-3, ax.get_position().y0, 12e-3, ax.get_position().height])
        cb = plt.colorbar(cax=cax, norm=norm_sos)
        cb.set_ticks([1500, 1520, 1540, 1560, 1580, 1600])
        cb.set_ticklabels(['1500', '1520', '1540', '1560', '1580', '1600'], fontsize=9, fontweight='bold')
        cb.set_label('$\mathbf{m \cdot s^{-1}}$', fontsize=12, fontweight='bold')

plt.savefig('figure_numerical_sos.png', dpi=128, bbox_inches='tight')
```
